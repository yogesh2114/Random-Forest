```python
import pandas as pd
import  numpy as np
from sklearn.model_selection import cross_val_score,KFold
from sklearn import preprocessing
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
fraud=pd.read_csv('Fraud.csv')
fraud
```

```python
fraud.shape
```

```python
le=preprocessing.LabelEncoder()
fraud['Urban']=le.fit_transform(fraud['Urban'])
fraud['Marital.Status']=le.fit_transform(fraud['Marital.Status'])
fraud['Undergrad']=le.fit_transform(fraud['Undergrad'])
```

```python
fraud
```

```python
#encoded the labels  as 0 = No and 1=Yes for urban
#encoded the labels as 0 = No and 1 = Yes for undergad
#encoded the labels as 0=divorced,1=married,2=single for  marital.status
```

```python
fraud['Taxable.Income']= pd.cut(x = fraud['Taxable.Income'],bins = [0,30000,100000], labels=['risk','safe'])
fraud
```

```python
fraud['Taxable.Income'].value_counts()
```

```python
fraud['Taxable.Income']=le.fit_transform(fraud['Taxable.Income'])
```

```python
fraud
```

```python
#encoded labels as 0=risk,1=safe for Taxable.Income
```

```python
x=fraud.drop(['Taxable.Income'],axis=1)
x
```

```python
y=fraud['Taxable.Income']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

#Bagging

```python
cart=DecisionTreeClassifier()
n_trees=15
```

```python
model=BaggingClassifier(base_estimator=cart,n_estimators=n_trees,random_state=41)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
result_cvr=cross_val_score(model,x_train,y_train,cv=10)
result_cvr.mean()
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

#Random forest

```python
model=RandomForestClassifier(n_estimators=500,max_features=3)
model.fit(x_train,y_train)
```
